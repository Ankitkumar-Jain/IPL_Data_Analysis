```python
teams = {'Royal Challengers Bangalore':'RCB','Kings XI Punjab':'KXIP','Delhi Daredevils':'DC','Mumbai Indians':'MI',
        'Kolkata Knight Riders':'KKR','Rajasthan Royals':'RR','Deccan Chargers':'SRH','Chennai Super Kings':'CSK',
        'Kochi Tuskers Kerala':'KTK','Pune Warriors':'PW','Sunrisers Hyderabad':'SRH','Gujarat Lions':'GL',
        'Rising Pune Supergiants':'PW','Rising Pune Supergiant':'PW','Delhi Capitals':'DC'}
```

```python
df['team1']=df['team1'].map(teams).fillna(df['team1'])
df['team2']=df['team2'].map(teams).fillna(df['team2'])
df['toss_winner']=df['toss_winner'].map(teams).fillna(df['toss_winner'])
df['winner'] = df['winner'].map(teams).fillna(df['winner'])
```

# Visualisation of Toss Won by Each Team

```python
teams = df.toss_winner.value_counts()
plt.bar(teams.index,teams.values)
plt.title("Toss Winning Team")
plt.xlabel('Teams')
plt.ylabel('No. of times toss won')
plt.show()
```

```python
df2['match_winner']=np.where((df2['toss_winner']==df2['winner']),df['toss_winner'],np.nan)
```

```python
df2
```

```python
match_won = df2.match_winner.value_counts()
```

```python
match_won
```

```python
df3=pd.concat([teams,match_won],axis=1)
```

```python
df3['team']=df3.index
df3
```

# Visualisation of Toss won by team also won the match

```python
df3.plot(kind='bar',x='team',title='Toss won vs Match won')
plt.ylabel('No. of times Toss won vs match won')
plt.show()
```
